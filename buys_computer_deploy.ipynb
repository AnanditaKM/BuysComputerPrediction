```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import pickle


computer = pd.read_csv("C:/Users/User/Downloads/buys_computer.csv")
computer
```

```python
label=LabelEncoder()
computer['age'] = label.fit_transform(computer['age'])
computer['income'] = label.fit_transform(computer['income']) 
computer['student'] = label.fit_transform(computer['student']) 
computer['credit_rating'] = label.fit_transform(computer['credit_rating']) 
computer['buys_computer'] = label.fit_transform(computer['buys_computer'])

computer
```

```python
x=computer.iloc[:,:-1].values
y=computer.iloc[:,-1].values
```

```python
x_train, x_test, y_train, ytest= train_test_split(x, y,train_size=0.8)
len(x_train)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred_dt = dtc.predict(x_train)
accuracy =metrics.accuracy_score(y_train, y_pred_dt)
print("akurasinya",accuracy)
```

```python
test_case=dtc.predict([[2,2,1,0]])
print(test_case)
```
